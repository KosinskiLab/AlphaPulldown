```python
from Bio import AlignIO

# Read the two MSAs
msa_ap = AlignIO.read("ap_msa_with_pairing.sto", "stockholm")
msa_af = AlignIO.read("af_msa.sto", "stockholm")

# Build dicts for each
seqs_ap = {record.id: str(record.seq) for record in msa_ap}
seqs_af = {record.id: str(record.seq) for record in msa_af}

# Build reverse lookup for the af MSA: sequence → ID
rev_af = {seq: rec_id for rec_id, seq in seqs_af.items()}

# Compare
not_found = 0
swapped = 0
for ap_id, ap_seq in seqs_ap.items():
    af_id = rev_af.get(ap_seq)
    if af_id is None:
        # In case no match is found (shouldn't happen if the sequences truly match)
        print(f"{ap_id} not found in af_msa")
        not_found += 1
    else:
        # If IDs differ, they're swapped
        if ap_id != af_id:
            print(f"{ap_id} <--> {af_id}")
            swapped += 1
print(f"{not_found} unique sequences in AP compared to AF")
print(f"{swapped} sequences are swapped in AP compared to AF")
```
